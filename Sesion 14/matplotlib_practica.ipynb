# Punto 0:

```python
# importar pandas
import pandas as pd
```

```python
# datos de RRSS 2020
fbk = ['Facebook', 2449, True, 2006]
twt = ['Twitter', 339, False, 2006]
ig = ['Instagram', 1000, True, 2010]
yt = ['YouTube', 2000, False, 2005]
lkn = ['LinkedIn', 663, False, 2003]
wsp = ['WhatsApp', 1600, True, 2009]
```

```python
# lista con datos
lista_rrss = [fbk, twt, ig, yt, lkn, wsp]
```

```python
# crear dataframe a partir de listas
df_rrss=pd.DataFrame(lista_rrss,
            columns = ['Nombre', 'Cantidad', 'ES_FBK', 'Año'])
```

```python
df_rrss
```

```python
# matplotlib basics 101
```

```python
# importar matplotlib
import matplotlib.pyplot as plt
```

```python
# ejemplo super mega básico
x=[1,2,3,4,5]
y=[1,8,27,64,125]

plt.plot(x,y)
plt.show()
```

# Punto 1:

```python
# gráfico de líneas utilizando matplotlib
plt.figure(figsize=(8,5))
plt.plot(df_rrss["Nombre"], df_rrss["Cantidad"])
plt.title("Usuarios de Redes Sociales")
plt.xlabel("Red Social")
plt.ylabel("Cantidad (millones)")
plt.grid(True)
plt.show()
```

```python
# gráfico de puntos utilizando matplotlib
plt.figure(figsize=(8,5))
plt.scatter(df_rrss["Nombre"], df_rrss["Cantidad"])
plt.title("Usuarios de Redes Sociales")
plt.xlabel("Red Social")
plt.ylabel("Cantidad (millones)")
plt.grid(True)
plt.show()
```

```python
# EXTRA: cómo hacer lo mismo con métodos de un dataframe de pandas
df_rrss.plot(
    x="Nombre",
    y="Cantidad",
    kind="scatter",
    figsize=(8,5),
    title="Usuarios de Redes Sociales"
)
plt.show()
```

# Punto 2:

```python
# hacer gráficos de barras

plt.figure(figsize=(8,5))
plt.bar(df_rrss["Nombre"], df_rrss["Cantidad"])
plt.title("Usuarios por Red Social")
plt.xlabel("Red Social")
plt.ylabel("Cantidad (millones)")
plt.show()
```

```python
# gráfico de puntos utilizando matplotlib
plt.figure(figsize=(8,5))
plt.scatter(df_rrss["Nombre"], df_rrss["Cantidad"])
plt.title("Comparación de Usuarios")
plt.xlabel("Red Social")
plt.ylabel("Cantidad (millones)")
plt.show()
```

```python
# gráfico de barras ordenado
df_ordenado = df_rrss.sort_values(by="Cantidad", ascending=False)

plt.figure(figsize=(8,5))
plt.bar(df_ordenado["Nombre"], df_ordenado["Cantidad"])
plt.title("Redes Sociales Ordenadas")
plt.xlabel("Red Social")
plt.ylabel("Cantidad (millones)")
plt.show()
```

```python
# gráfico de barras ordenado y con color


# b: blue.
# g: green.
# r: red.
# c: cyan.
# m: magenta.
# y: yellow.
# k: black.
# w: white.

```

```python
df_ordenado = df_rrss.sort_values(by="Cantidad", ascending=False)

plt.figure(figsize=(8,5))
plt.bar(
    df_ordenado["Nombre"],
    df_ordenado["Cantidad"],
    color=['b','r','g','m','k','c']
)
plt.title("Redes Sociales Ordenadas por Usuarios")
plt.xlabel("Red Social")
plt.ylabel("Cantidad (millones)")
plt.show()
```

# Punto 3:

```python
# Hacer gráficas de pastel/tarta/torta/pie

plt.figure(figsize=(8,8))
plt.pie(
    df_rrss["Cantidad"],
    labels=df_rrss["Nombre"],
    autopct="%1.1f%%"
)

plt.title("Participación de Redes Sociales")
plt.show()
```

```python
# gráfico de puntos utilizando matplotlib utilizando colores
```

```python
lista_col_rrss=['#3b5998', '#FF0000', '#25d366', '#8a3ab9', '#0e76a8', '#00acee']
```

```python
plt.figure(figsize=(8,5))

plt.scatter(
    df_rrss["Nombre"],
    df_rrss["Cantidad"],
    c=lista_col_rrss,
    s=150
)

plt.title("Usuarios de Redes Sociales")
plt.xlabel("Red Social")
plt.ylabel("Cantidad")
plt.grid(True)
plt.show()
```

```python
# gráfico de torta utilizando matplotlib por Color Corporativo
```

```python
plt.figure(figsize=(8,8))

plt.pie(
    df_rrss["Cantidad"],
    labels=df_rrss["Nombre"],
    colors=lista_col_rrss,
    autopct="%1.1f%%"
)

plt.title("Participación de Redes Sociales por Color Corporativo")
plt.show()
```
