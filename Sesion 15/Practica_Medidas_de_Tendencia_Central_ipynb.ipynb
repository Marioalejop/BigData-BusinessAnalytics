## PRACTICA DE MEDIDAS DE TENDENCIA CENTRAL

## Medidas de tendencia central

### Media = $$ \frac{1}{N} \sum_{i=1}^N x_i  $$

### Mediana(impar) = $$ x_{(n+1)/2}^{\text{ordered}} $$

### Mediana(par) = $$ \frac{x_{n/2}^{\text{ordered}} + x_{n/2+1}^{\text{ordered}}}{2} $$

### Moda = $$ x_k $$ donde $$ \text{Freq}(x_k) = \max{(\text{Freq}(x_i))} $$

```python
# Filtra los datos para aislar los modelos Audi Q7 y genera un histograma de sus precios diferenciando por año de producción.
```

```python
Q7_df = df[(df['model_name'] == 'Q7') & (df['manufacturer_name'] == 'Audi')]
sns.histplot(Q7_df, x = 'price_usd', hue = 'year_produced')
```

Esta gráfica se centra en el modelo Audi Q7. Muestra la distribución de precios de los Q7, diferenciando por el año de producción. Generalmente, se espera ver que los modelos de años más recientes tienen precios más elevados y los modelos más antiguos, precios menores.


```python
A6_df = df[(df['model_name'] == 'A6') & (df['manufacturer_name'] == 'Audi')]
sns.histplot(A6_df, x = 'price_usd', hue = 'year_produced')
```

Esta gráfica se centra en el modelo Audi A6 y muestra la distribución de precios de los A6, diferenciando por el año de producción. Similar al Q7, se espera una tendencia con precios variando por año de producción, donde modelos más recientes tienden a ser más caros.


```python
# Grafico de pastel con las marcas de los carros
```

```python
manufacturer_distribution = df['manufacturer_name'].value_counts()

plt.figure(figsize=(12, 10))
plt.pie(manufacturer_distribution, labels=manufacturer_distribution.index, autopct='%1.1f%%', startangle=90, pctdistance=0.85)
plt.title('Distribución de Marcas de Vehículos Más Vendidas')
plt.axis('equal')
plt.show()
```

#### Observaciones Clave de la Distribución de Marcas

El gráfico de pastel muestra claramente que algunas marcas tienen una presencia significativamente mayor en el dataset que otras. Por ejemplo, `Volkswagen` es la marca más predominante.
