Actividad 5

```python
# Extraer datos de Pandas a una lista
```

```python
lista_tarifas = df_titanic['Fare'].head(20).tolist()
```

```python
lista_tarifas
```

Actividad 6

```python
# List Comprehensions (Listas de comprensión) para limpieza
```

```python
tarifas_redondeadas = [round(tarifa, 2) for tarifa in lista_tarifas]
```

```python
tarifas_redondeadas
```

Actividad 7

```python
# Convertir listas a Arrays de NumPy
```

```python
array_tarifas = np.array(tarifas_redondeadas)
```

```python
array_tarifas
```

Actividad 8

```python
# Matemáticas vectorizadas con NumPy
```

```python
tarifas_impuesto = array_tarifas * 1.15
```

```python
tarifas_impuesto
```

Actividad 9

```python
# Funciones estadísticas de NumPy
```

```python
print("Máximo:", np.max(array_tarifas))
print("Mínimo:", np.min(array_tarifas))
print("Desviación estándar:", np.std(array_tarifas))
```

Actividad 10

```python
#  Lógica condicional con np.where
```

```python
edades = df_titanic['Age'].dropna().values
```

```python
clasificacion = np.where(edades >= 18, 'Mayor de edad', 'Menor de edad')

for i in range(len(edades)):
    print(f'{clasificacion[i]}: {edades[i]} años')
```

Actividad 17

```python
# Tratamiento de valores nulos (Imputación)
```

```python
mediana_edad = df_titanic['Age'].median()
```
