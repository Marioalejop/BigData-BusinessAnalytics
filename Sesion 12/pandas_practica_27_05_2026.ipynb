```python
# importar pandas
```

```python
import pandas as pd
```

```python
# datos de RRSS 2020
```

```python
fbk = ['Facebook', 2449, True, 2006]
twt = ['Twitter', 339, False, 2006]
ig = ['Instagram', 1000, True, 2010]
yt = ['YouTube', 2000, False, 2005]
lkn = ['LinkedIn', 663, False, 2003]
wsp = ['WhatsApp', 1600, True, 2009]
```

```python
# crear dataframe a partir de listas
```

```python
lista_rrss = [fbk, twt, ig, yt, lkn, wsp]
```

```python
df_rrss = pd.DataFrame(lista_rrss,
                       columns = ['Nombre', 'Cantidad', 'ES_FBK', 'Año'])
```

```python
df_rrss
```

```python
# crear dataframe vacio
```

```python
df_vacio = pd.DataFrame(columns = ['Nombre', 'Cantidad', 'ES_FBK', 'Año'])
```

```python
df_vacio
```

```python
new_row = pd.DataFrame([{'Nombre' : 'Facebook',
                            'Cantidad' : 2449,
                            'ES_FBK' : True,
                            'Año' : 2006}])
df_vacio = pd.concat([df_vacio, new_row], ignore_index=True)
```

```python
df_vacio
```

```python
new_row = pd.DataFrame([{'Nombre' : 'Instagram',
                            'Cantidad' : 1000,
                            'ES_FBK' : True,
                            'Año' : 2010}])
df_vacio = pd.concat([df_vacio, new_row], ignore_index=True)
```

```python
df_vacio
```
