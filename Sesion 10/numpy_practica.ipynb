# Punto 3:

```python
# Como seleccionar elements
```

```python
datos = np.array([
    [10, 20, 30],
    [40, 50, 60],
    [70, 80, 90]
])
```

```python
datos
```

```python
# Seleccionar un elemento
```

```python
datos[1,1]
```

```python
datos[2,2]
```

```python
# seleccionar el primer elemento
```

```python
datos[0,0]
```

```python
# seleccionar algún otro elemento
```

```python
datos[0,2]
```

```python
# Seleccionar varios elementos
```

```python
datos[0:2]
```

```python
# seleccionar por fila
```

```python
datos[1]
```

```python
# seleccionar por columna
```

```python
datos[:,1]
```

```python
# Utilizar filter
```

```python
numeros = np.array([5, 12, 18, 25, 30, 8])
```

```python
numeros
```

```python
# condición mayores de 15
```

```python
numeros > 15
```

```python
# conidicion para menores de 20
```

```python
numeros < 20
```

```python
# filtrar menores de 20
```

```python
numeros[numeros < 20]
```

```python
# filtrar mayores o iguales a 20
```

```python
numeros[numeros >= 20]
```
