```python
# Tomar solo películas
```

```python
peliculas = df[df['type'] == 'Movie']
```

```python
peliculas[['title','duration']].head(10)
```

```python
# Limpiar duración
```

```python
duraciones = []

for d in peliculas['duration'].head(10):
    numero = int(d.replace(' min',''))
    duraciones.append(numero)
```

```python
duraciones
```

```python
# Crear ndarray
```

```python
duraciones_np = np.array(duraciones)
```

```python
duraciones_np
```

```python
# Buffer del 5%
```

```python
buffer_5 = duraciones_np * 1.05
```

```python
buffer_5
```

```python
# Duración ideal
```

```python
duracion_ideal = np.array([100]*10)
```

```python
duracion_ideal
```

```python
# Desviación
```

```python
desviacion = duraciones_np - duracion_ideal
```

```python
desviacion
```

```python
# Estadísticas
```

```python
print("Promedio:", np.mean(duraciones_np))
print("Máximo:", np.max(duraciones_np))
print("Mínimo:", np.min(duraciones_np))
```
